```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pathlib
```

```python
## Self explanatory
save_plots = False
```

```python
output_dir = './data/'

df_perfect = pd.read_hdf(output_dir + 'pseudoexperiments_perfect.h5', 'LHCb_Pseudo')
df_nocross = pd.read_hdf(output_dir + 'pseudoexperiments_nocross.h5', 'LHCb_Pseudo')
df_current = pd.read_hdf(output_dir + 'pseudoexperiments_current.h5', 'LHCb_Pseudo')
```

```python
for df in [df_perfect, df_nocross, df_current]:
    df['PULL'] = (df['g'] - 1.226)/df['g_ERR']
```

```python
plt.hist(df_perfect['g'], bins=100, histtype='step', range=(0.5,1.5), linewidth=1.5, label='No smearing')
plt.hist(df_nocross['g'], bins=100, histtype='step', range=(0.5,1.5), linewidth=1.5, label='No ghost')
plt.hist(df_current['g'], bins=100, histtype='step', range=(0.5,1.5), linewidth=1.5, label='With ghost')
plt.legend(frameon=False, loc='upper left')
plt.xlabel('$g$')
plt.ylabel('Pseudoexperiments')
plt.show()
```

```python
plt.hist(df_perfect['g_ERR'], bins=100, histtype='step', range=(0.052, 0.085), linewidth=1.5, label='No smearing')
plt.hist(df_nocross['g_ERR'], bins=100, histtype='step', range=(0.052, 0.085), linewidth=1.5, label='No ghost')
plt.hist(df_current['g_ERR'], bins=100, histtype='step', range=(0.052, 0.085), linewidth=1.5, label='With ghost')
plt.legend(frameon=False, loc='upper right')
plt.xlabel('$\sigma_g$')
plt.ylabel('Pseudoexperiments')
plt.show()
```

```python
plt.hist(df_perfect['PULL'], bins=100, histtype='step', range=(-10,4), linewidth=1.5, label='No smearing')
plt.hist(df_nocross['PULL'], bins=100, histtype='step', range=(-10,4), linewidth=1.5, label='No ghost')
plt.hist(df_current['PULL'], bins=100, histtype='step', range=(-10,4), linewidth=1.5, label='With ghost')
plt.legend(frameon=False, loc='upper left')
plt.xlabel('$g$ pull')
plt.ylabel('Pseudoexperiments')
plt.show()
```
